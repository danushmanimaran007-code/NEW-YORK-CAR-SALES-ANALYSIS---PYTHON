```python
def transmissionChange(x):
    gear = [0, 0]
    if("Auto" in x.values[0] or "A/T" in x.values[0]):
        gear[0] = 1
    if("Manual" in x.values[0] or "M/T" in x.values[0] or "Sequential" in x.values[0]):
        gear[1] = 1
    return gear
```

```python
df_transmission = df_cars[["Transmission"]].apply(transmissionChange, axis = 1, result_type = "expand")
```
